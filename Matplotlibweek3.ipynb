```python
x1 = np.linspace(0.0, 5.0, 100)
x2 = np.linspace(0.0, 2.0, 100)
y1 = np.cos(2*np.pi*x1)
y2 = np.cos(2*np.pi*x2)
```

```python
x1
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1)
fig.suptitle('Multiple Subplots')

ax1.plot(x1, y1, 'ro-')
ax1.set_ylabel('Nilai cos(x1)')

ax2.plot(x2, y2, 'g.-')
ax2.set_ylabel('Nilai cos(x2)')

ax2.set_xlabel('Nilai x')
plt.show()
```

```python
fig, (bx1, bx2) = plt.subplots(2, 1)
fig.suptitle('$Multiple Subplots$')

bx1.plot(x1, y1, 'b*')
bx1.set_ylabel('Nilai $cos(x1)$')

bx2.plot(x2, y2, 'y.-')
bx2.set_ylabel('Nilai $cos(x2)$')

bx2.set_xlabel('Nilai $x$')
plt.show()
```

```python
plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'b*-')
plt.title('Multiple Subplots')
plt.ylabel('Nilai cos(x_1)')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r+-')
plt.ylabel('Nilai cos(x_2)')

plt.xlabel('Nilai x')
plt.show()
```

```python
mu, sigma = 100, 15
a = mu + sigma *np.random.randn(10000)
a
```

```python
a.shape
```

```python
plt.hist(a, bins=50, facecolor='y', alpha=0.75)
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Histogram')
plt.text(45, 500, r'$\mu=100,\ \sigma=15$')
plt.grid()
plt.show()
```

```python
plt.hist(a, bins=100, facecolor='r', alpha=0.50)
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Histogram')
plt.text(45, 500, '$\mu=100,\ \sigma=15$')
plt.grid()
plt.show()
```
