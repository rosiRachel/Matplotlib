```python
e = np.linspace(0, 5, 100)
e
```

```python
fig, ax = plt.subplots()

ax.plot(e, e, label='linear')
ax.plot(e, e**2, label='quadratic')
ax.plot(e, e**3, label='cubic')

ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('Simple Plot')
ax.legend()
```
