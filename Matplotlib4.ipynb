```python
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np
print(matplotlib.__version__)
print(np.__version__)
```

```python
a_data = [25, 45, 55, 125, 225]
a_kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(a_kategori, a_data)
plt.grid()
plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')
plt.show()
```

```python
b_data = [25, 35, 55, 75, 125]
b_kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(b_kategori, b_data, color='red', alpha=0.25)
plt.grid(linestyle='--', linewidth=2, axis='y', alpha=0.50)
plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')
plt.show()
```

```python
plt.bar(b_kategori, b_data, color='green', alpha=0.75)
plt.grid(linestyle='--', linewidth=4, axis='x', alpha=0.25)
plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')
plt.show()
```

```python
c_data = [30, 45, 55, 75, 125]
c_kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.barh(c_kategori, c_data)
plt.grid(linestyle='--', linewidth=1, axis='x', alpha=0.75)
plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Horisontal Bar Plot')
plt.show()
```

```python
plt.barh(c_kategori[::-1], c_data[::-1])
plt.grid(linestyle='--', linewidth=1, axis='x', alpha=0.75)
plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Horisontal Bar Plot')
plt.show()
```

```python
d_data = [25, 85, 75, 40, 60]
e_data = [40, 35, 20, 55, 10]
d_kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(d_kategori, d_data, label='Data 1')
plt.bar(d_kategori, e_data, label='Data 2', bottom=d_data)
plt.legend()
plt.grid(linestyle='--', linewidth=1, axis='y', alpha=0.75)
plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Stacked Bar Plot') #Diagram batang bertumpuk
plt.show()
```

```python
f_data = [10, 85, 90, 40, 60]
g_data = [40, 65, 20, 55, 10]
e_kategori = ['A', 'B', 'C', 'D', 'E']
h = np.arange(len(e_kategori))
h
```

```python
width = 0.35
plt.bar(h-width/2, f_data, width, label='Data 1')
plt.bar(h+width/2, g_data, width, label='Data 2')
plt.xticks(h, e_kategori)
plt.grid(linestyle='--', linewidth=1, axis='y', alpha=0.75)
plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Gouped Bar Plot')
plt.legend()
plt.show()
```

```python
i_data = [215, 130, 245, 210, 225]
j_kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.pie(i_data, labels=j_kategori, autopct='%1.1f%%', startangle=90)
plt.title('Contoh Pie Plot')
plt.show()
```

```python
k_data = [215, 130, 245, 210, 225]
l_kategori = ['A', 'B', 'C', 'D', 'E']
a_warna = ['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue']
```

```python
plt.pie(k_data, labels=l_kategori, colors=a_warna, autopct='%1.1f%%', startangle=90)
plt.legend()
plt.title('Contoh Pie Plot')
plt.show()
```

```python
m_data = [215, 130, 245, 210, 225]
n_kategori = ['A', 'B', 'C', 'D', 'E']
b_warna = ['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue']
explode_var = [0., 0., 0.2, 0., 0.08]
```

```python
plt.pie(m_data, labels=n_kategori, colors=b_warna, explode=explode_var, autopct='%1.1f%%', startangle=90)
plt.title('Contoh Exploded Pie Plot')
plt.show()
```

```python
data_uas = [['Bejo', 70], ['Tejo', 83], ['Cecep', 62], ['Wati', 74], ['Karti', 71]]
data_uas
```
